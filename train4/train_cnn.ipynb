```python
# train_tl_full_aug_label_with_fallback.py
import os, glob, time, random
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, Input
from tensorflow.keras.applications import MobileNetV2
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import gc
import importlib

# ---------------- Settings (แก้สำหรับ dataset ขนาดใหญ่) ----------------
RESIZED = (144, 144)   # (W,H)
RANDOM_STATE = 42
BATCH_SIZE = 64
EPOCHS = 150
VERBOSE = 1
IMG_DIR = "./../data/Right/Image144x144"
ANN_DIR = "./../data/Right/Annotation144x144"
MODEL_SAVE_DIR = "./models_tl"
os.makedirs(MODEL_SAVE_DIR, exist_ok=True)

# GPU memory growth
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Found {len(gpus)} GPU(s), memory growth enabled")
    except RuntimeError as e:
        print(e)

np.random.seed(RANDOM_STATE)
random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

# try import tensorflow_addons.image (only needed if available)
_tfa_image = None
try:
    tfa_mod = importlib.import_module("tensorflow_addons.image")
    _tfa_image = tfa_mod
    print("tensorflow_addons.image found -> will use tfa.rotate when rotating.")
except Exception:
    _tfa_image = None
    print("tensorflow_addons.image not found -> will use fallback rotate via tf.raw_ops.ImageProjectiveTransformV3.")

# ---------------- Helpers ----------------
def read_annotation_raw(txt_path):
    if not os.path.exists(txt_path):
        return None
    with open(txt_path, 'r', encoding='utf-8') as f:
        content = f.read().strip()
        tokens = content.split()
        vals = []
        for t in tokens:
            try: vals.append(float(t))
            except: pass
        if len(vals) >= 3:
            return vals[0], vals[1], vals[2]
    return None

def load_images_and_labels(img_dir, ann_dir, size=RESIZED, use_clahe=True):
    imgs_paths = []
    for ext in ('*.png','*.jpg','*.jpeg','*.bmp'):
        imgs_paths += glob.glob(os.path.join(img_dir, ext))
    imgs_paths = sorted([p for p in imgs_paths if not os.path.basename(p).lower().startswith("annotated_")])
    X_resized, Y_resized, Y_orig, orig_shapes, filenames = [], [], [], [], []
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    skipped = 0
    for p in imgs_paths:
        name = os.path.basename(p)
        base = os.path.splitext(name)[0]
        img = cv2.imread(p, cv2.IMREAD_GRAYSCALE)
        if img is None:
            skipped += 1
            continue
        h, w = img.shape[:2]
        orig_shapes.append((w,h))
        filenames.append(base)

        ann = read_annotation_raw(os.path.join(ann_dir, base + ".txt"))
        if ann is None:
            x_px = y_px = r_px = None
        else:
            x_raw, y_raw, r_raw = ann
            if 0.0 <= x_raw <= 1.0 and 0.0 <= y_raw <= 1.0:
                x_px = x_raw * w
                y_px = y_raw * h
                r_px = r_raw * min(w,h)
            else:
                x_px, y_px, r_px = x_raw, y_raw, r_raw
        Y_orig.append([x_px, y_px, r_px])

        im_proc = img.copy()
        if use_clahe:
            im_proc = clahe.apply(im_proc)
        imr = cv2.resize(im_proc, (size[0], size[1]), interpolation=cv2.INTER_AREA)
        imr = imr.astype(np.float32)/255.0
        imr = imr[..., np.newaxis]
        X_resized.append(imr)

        if x_px is None:
            Y_resized.append([np.nan,np.nan,np.nan])
        else:
            scale_x = size[0]/w
            scale_y = size[1]/h
            scale_r = (scale_x+scale_y)/2.0
            Y_resized.append([x_px*scale_x, y_px*scale_y, r_px*scale_r])
    print(f"Loaded {len(X_resized)} images, skipped {skipped}")
    return (np.array(X_resized, dtype=np.float32),
            np.array(Y_resized, dtype=np.float32),
            np.array(Y_orig, dtype=np.float32),
            orig_shapes, filenames)

# ---------------- Rotate fallback util ----------------
@tf.function
def _rotate_image_with_fallback(image, angle):
    """
    Rotate image by angle (radians). If tensorflow_addons.image.rotate available -> use it.
    Otherwise use tf.raw_ops.ImageProjectiveTransformV3 as fallback.
    image: [H,W,C], angle: scalar radians
    """
    if _tfa_image is not None:
        # use tfa.rotate (wrap call in tf.numpy_function style isn't necessary since tfa is TF op)
        return _tfa_image.rotate(image, angles=angle, fill_mode='reflect')
    else:
        # fallback: build transform for rotation about image center
        h = tf.cast(tf.shape(image)[0], tf.float32)
        w = tf.cast(tf.shape(image)[1], tf.float32)
        cx = (w - 1.0) / 2.0
        cy = (h - 1.0) / 2.0

        cos_a = tf.cos(angle)
        sin_a = tf.sin(angle)

        # compute translation to keep rotation about center
        tx = cx - (cos_a * cx - sin_a * cy)
        ty = cy - (sin_a * cx + cos_a * cy)

        # transforms param for ImageProjectiveTransformV3: [a0,a1,a2,a3,a4,a5,a6,a7]
        transforms = tf.stack([cos_a, -sin_a, tx, sin_a, cos_a, ty, 0.0, 0.0])
        transforms = tf.reshape(transforms, (1, 8))
        img_exp = tf.expand_dims(image, 0)  # [1,H,W,C]
        out_h = tf.shape(image)[0]
        out_w = tf.shape(image)[1]
        out = tf.raw_ops.ImageProjectiveTransformV3(
            images=img_exp,
            transforms=transforms,
            output_shape=tf.cast(tf.stack([out_h, out_w]), tf.int32),
            interpolation="BILINEAR",
            fill_value=0.0
        )
        out = tf.squeeze(out, axis=0)
        return out

# ---------------- Augmentation (tf ops ทั้งหมด) ----------------
# note: image shape assumed (H,W,3) float32 in [0,1], label = [cx,cy,r] in pixel coords (resized image pixels)
@tf.function
def tf_augment_strong(image, label):
    # brightness & contrast
    image = tf.image.random_brightness(image, max_delta=0.15)
    image = tf.image.random_contrast(image, lower=0.85, upper=1.15)

    h = tf.cast(tf.shape(image)[0], tf.float32)
    w = tf.cast(tf.shape(image)[1], tf.float32)

    # --- Rotation (update label) ---
    angle = tf.random.uniform([], -0.2, 0.2)  # radians
    image = _rotate_image_with_fallback(image, angle)

    cx = label[0]
    cy = label[1]
    cx_c = (w - 1.0) / 2.0
    cy_c = (h - 1.0) / 2.0
    cos_a = tf.cos(angle)
    sin_a = tf.sin(angle)
    x_shifted = cx - cx_c
    y_shifted = cy - cy_c
    x_rot = cos_a * x_shifted - sin_a * y_shifted
    y_rot = sin_a * x_shifted + cos_a * y_shifted
    cx_rot = x_rot + cx_c
    cy_rot = y_rot + cy_c
    r = label[2]

    # --- Random Zoom (scale) (update label and radius) ---
    do_zoom = tf.random.uniform([]) < 0.5
    def apply_zoom():
        scale = tf.random.uniform([], 0.92, 1.08)
        new_h = tf.cast(tf.cast(h, tf.float32) * scale, tf.int32)
        new_w = tf.cast(tf.cast(w, tf.float32) * scale, tf.int32)
        img_resized = tf.image.resize(image, [new_h, new_w], method='bilinear')
        img_cropped = tf.image.resize_with_crop_or_pad(img_resized, tf.cast(h, tf.int32), tf.cast(w, tf.int32))
        cx_s = cx_rot * scale
        cy_s = cy_rot * scale
        offset_x = (tf.cast(new_w, tf.float32) - w) / 2.0
        offset_y = (tf.cast(new_h, tf.float32) - h) / 2.0
        cx_s = cx_s - offset_x
        cy_s = cy_s - offset_y
        r_s = r * scale
        return img_cropped, cx_s, cy_s, r_s
    def skip_zoom():
        return image, cx_rot, cy_rot, r
    image, cx_z, cy_z, r_z = tf.cond(do_zoom, apply_zoom, skip_zoom)

    # --- Random Translation (roll) (update label) ---
    do_trans = tf.random.uniform([]) < 0.5
    max_trans_frac = 0.06
    def apply_trans():
        tx = tf.random.uniform([], -max_trans_frac, max_trans_frac)
        ty = tf.random.uniform([], -max_trans_frac, max_trans_frac)
        tx_px = tf.cast(tf.round(tx * w), tf.int32)
        ty_px = tf.cast(tf.round(ty * h), tf.int32)
        img_t = tf.roll(image, shift=[ty_px, tx_px], axis=[0,1])
        cx_t = cx_z + tf.cast(tx_px, tf.float32)
        cy_t = cy_z + tf.cast(ty_px, tf.float32)
        return img_t, cx_t, cy_t
    def skip_trans():
        return image, cx_z, cy_z
    image, cx_t, cy_t = tf.cond(do_trans, apply_trans, skip_trans)

    # Clip labels to image bounds (optional)
    cx_t = tf.clip_by_value(cx_t, 0.0, w - 1.0)
    cy_t = tf.clip_by_value(cy_t, 0.0, h - 1.0)
    r_z = tf.maximum(1.0, tf.minimum(r_z, tf.reduce_max([w,h])))
    new_label = tf.stack([cx_t, cy_t, r_z], axis=0)
    image = tf.clip_by_value(image, 0.0, 1.0)
    return image, new_label

@tf.function
def convert_to_3channel(image, label):
    image = tf.repeat(image, 3, axis=-1)
    return image, label

# ---------------- Build TL model ----------------
def build_tl_small(input_shape=(RESIZED[1], RESIZED[0],3)):
    inp = Input(shape=input_shape)
    base = MobileNetV2(input_tensor=inp, include_top=False, weights='imagenet')
    base.trainable = False
    x = base.output
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.1)(x)
    out = layers.Dense(3, activation='linear', dtype='float32')(x)
    return models.Model(inp,out)

# ---------------- Load & prepare data ----------------
print("Loading data...")
X_resized, Y_resized, Y_orig, orig_shapes, filenames = load_images_and_labels(IMG_DIR, ANN_DIR)

mask = ~np.isnan(Y_resized).any(axis=1)
X_labeled = X_resized[mask]
y_labeled = Y_resized[mask]
orig_shapes_labeled = [orig_shapes[i] for i in range(len(orig_shapes)) if mask[i]]
filenames_labeled = [filenames[i] for i in range(len(filenames)) if mask[i]]
Y_orig_labeled = np.array(Y_orig)[mask]  # original pixel GT for labeled subset
print(f"Annotated images: {len(y_labeled)}/{len(Y_resized)}")

# train/val split WITH indices เพื่อ mapping กลับ
indices = np.arange(len(X_labeled))
X_tr, X_val, y_tr, y_val, idx_tr, idx_val = train_test_split(
    X_labeled, y_labeled, indices, test_size=0.15, random_state=RANDOM_STATE
)
print("Train/Val sizes:", X_tr.shape, X_val.shape)
val_orig_shapes = [orig_shapes_labeled[i] for i in idx_val]
val_filenames = [filenames_labeled[i] for i in idx_val]
y_val_orig = Y_orig_labeled[idx_val]

# free some memory
del X_resized, Y_resized, X_labeled
gc.collect()

# ---------------- tf.data pipeline ----------------
print("Creating efficient data pipeline...")

train_ds = tf.data.Dataset.from_tensor_slices((X_tr, y_tr))
train_ds = train_ds.shuffle(2048, seed=RANDOM_STATE, reshuffle_each_iteration=True)
train_ds = train_ds.map(convert_to_3channel, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.map(tf_augment_strong, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.batch(BATCH_SIZE)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((X_val, y_val))
val_ds = val_ds.map(convert_to_3channel, num_parallel_calls=tf.data.AUTOTUNE)
val_ds = val_ds.batch(BATCH_SIZE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)

print("✅ Data pipeline ready (memory efficient!)")

# free training arrays (we have datasets now)
del X_tr, y_tr
gc.collect()

# ---------------- Build & compile model ----------------
print("Building model...")
model = build_tl_small()
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss='mse',
    metrics=['mae']
)
model.summary()

# ---------------- Callbacks ----------------
ckpt_path = os.path.join(MODEL_SAVE_DIR, "tl_train_best.keras")
cb_list = [
    callbacks.ModelCheckpoint(
        ckpt_path,
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    ),
    callbacks.EarlyStopping(
        monitor='val_loss',
        patience=30,
        restore_best_weights=True,
        verbose=1
    ),
    callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=8,
        min_lr=1e-6,
        verbose=1
    ),
    callbacks.TensorBoard(
        os.path.join(MODEL_SAVE_DIR, "logs", f"tl_{int(time.time())}")
    )
]

# ---------------- Train ----------------
print("\n" + "="*50)
print("Starting training...")
print("="*50)
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=cb_list,
    verbose=VERBOSE
)

# ---------------- Save final model ----------------
final_model_path = os.path.join(MODEL_SAVE_DIR, "tl_trained_final.keras")
model.save(final_model_path)
print(f"\n✅ Saved final model to: {final_model_path}")

# ---------------- Predict & evaluate ----------------
print("\nEvaluating on validation set...")

# prepare eval dataset from X_val only
eval_ds = tf.data.Dataset.from_tensor_slices(X_val)
eval_ds = eval_ds.map(lambda x: tf.repeat(x, 3, axis=-1), num_parallel_calls=tf.data.AUTOTUNE)
eval_ds = eval_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

preds_resized = model.predict(eval_ds, verbose=1)

def resize_preds_to_orig(preds, orig_shapes, size=RESIZED):
    out = []
    for i, p in enumerate(preds):
        w, h = orig_shapes[i]
        sx = w / size[0]
        sy = h / size[1]
        sr = (sx + sy) / 2
        out.append([p[0]*sx, p[1]*sy, p[2]*sr])
    return np.array(out)

preds_orig = resize_preds_to_orig(preds_resized, val_orig_shapes)

# metrics: compare preds_orig with y_val_orig (both in original pixel coords)
r2_cx = r2_score(y_val_orig[:,0], preds_orig[:,0])
r2_cy = r2_score(y_val_orig[:,1], preds_orig[:,1])
r2_r = r2_score(y_val_orig[:,2], preds_orig[:,2])
mae_cx = mean_absolute_error(y_val_orig[:,0], preds_orig[:,0])
mae_cy = mean_absolute_error(y_val_orig[:,1], preds_orig[:,1])
mae_r = mean_absolute_error(y_val_orig[:,2], preds_orig[:,2])
rmse_r = np.sqrt(mean_squared_error(y_val_orig[:,2], preds_orig[:,2]))
euclid_errs = np.sqrt((y_val_orig[:,0]-preds_orig[:,0])**2 +
                      (y_val_orig[:,1]-preds_orig[:,1])**2)

print("\n" + "="*60)
print("📊 VALIDATION SET EVALUATION RESULTS")
print("="*60)
print(f"R² Score:")
print(f"  • Center X: {r2_cx:.4f}")
print(f"  • Center Y: {r2_cy:.4f}")
print(f"  • Radius:   {r2_r:.4f}")
print(f"\nMean Absolute Error (MAE):")
print(f"  • Center X: {mae_cx:.2f} px")
print(f"  • Center Y: {mae_cy:.2f} px")
print(f"  • Radius:   {mae_r:.2f} px")
print(f"\nRMSE Radius: {rmse_r:.2f} px")
print(f"\nEuclidean Distance Error:")
print(f"  • Mean:   {euclid_errs.mean():.2f} px")
print(f"  • Median: {np.median(euclid_errs):.2f} px")
print(f"  • Max:    {euclid_errs.max():.2f} px")
print(f"  • 95th percentile: {np.percentile(euclid_errs, 95):.2f} px")
print("="*60)

# save CSV
rows = []
for i, fname in enumerate(val_filenames):
    rows.append({
        "file": fname,
        "gt_cx": float(y_val_orig[i,0]),
        "gt_cy": float(y_val_orig[i,1]),
        "gt_r": float(y_val_orig[i,2]),
        "pred_cx": float(preds_orig[i,0]),
        "pred_cy": float(preds_orig[i,1]),
        "pred_r": float(preds_orig[i,2]),
        "euclid_err": float(euclid_errs[i]),
        "r_err": float(abs(y_val_orig[i,2]-preds_orig[i,2]))
    })

df = pd.DataFrame(rows)
csv_path = os.path.join(MODEL_SAVE_DIR, "eval_summary_tl.csv")
df.to_csv(csv_path, index=False)
print(f"\n✅ Saved evaluation CSV to: {csv_path}")

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.title('Training History - Loss')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['mae'], label='Train MAE')
plt.plot(history.history['val_mae'], label='Val MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.title('Training History - MAE')
plt.grid(True)

plt.tight_layout()
plot_path = os.path.join(MODEL_SAVE_DIR, "training_history.png")
plt.savefig(plot_path, dpi=150)
print(f"✅ Saved training plot to: {plot_path}")

print("\n🎉 Training completed successfully!")
print(f"📁 All results saved in: {MODEL_SAVE_DIR}")
```
